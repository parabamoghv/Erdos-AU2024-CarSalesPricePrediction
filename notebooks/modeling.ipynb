```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
file_loc = "..\data\processed\Cat_details_v3_cleaned_entire.csv"
df = pd.read_csv(file_loc)
#The data source is the following https://www.kaggle.com/datasets/nehalbirla/vehicle-dataset-from-cardekho
```

```python
df = df.drop("Unnamed: 0", axis=1)
df.info()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
df_train.sample(5)
```

```python
## import the LinearRegression object
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
continuous_features = ["selling_price", "km_driven", "mileage", "engine", "max_power"]

simple_lr = LinearRegression()
scaler = StandardScaler()

scaler.fit(df_train[continuous_features])
train_scale = scaler.transform(df_train[continuous_features])

train_scale.shape

simple_lr.fit(train_scale[:,1:], train_scale[:,0])
```

```python
## look at coef and intercept
simple_lr.coef_, simple_lr.intercept_
```

```python
## Make a prediction
test_scale = scaler.transform(df_test[continuous_features])

simple_lr_pred = simple_lr.predict(test_scale[:,1:])
mse = mean_squared_error(test_scale[:,0], simple_lr_pred)
print("Mean Squared Error:", mse)
```
